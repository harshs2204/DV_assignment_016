```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
iris_data = pd.read_csv(url, names=columns)

# Basic Plot 1: Histogram
plt.figure(figsize=(10, 6))
plt.hist(iris_data['SepalLength'], bins=15, color='#fa5a9a', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Basic Plot 2: Bubble Plot
plt.figure(figsize=(10, 6))
plt.scatter(iris_data['SepalLength'], iris_data['SepalWidth'], s=iris_data['PetalLength']*30, alpha=0.5, c='purple')
plt.title('Bubble Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid()
plt.show()
```

```python
# Basic Plot 3: Box Plot
plt.figure(figsize=(10, 6))
iris_data.boxplot(column='SepalLength', by='Species', grid=False)
plt.title('Box Plot of Sepal Length by Species')
plt.suptitle('')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.grid(axis='y')
plt.show()
```

```python
# Basic Plot 4: Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(iris_data['PetalLength'], iris_data['PetalWidth'], color='orange')
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid()
plt.show()
```

```python
# Basic Plot 5: Heatmap
# Calculate the correlation matrix
correlation_matrix = iris_data.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='summer', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Heatmap of Feature Correlation')
plt.show()
```
