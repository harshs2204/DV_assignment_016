```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Sample Dataset
data = pd.DataFrame({
    'Platform': ['Instagram', 'Facebook', 'Twitter', 'TikTok', 'LinkedIn'],
    'Likes': [4500, 3500, 1500, 5000, 2000],
    'Comments': [1500, 1000, 700, 2000, 500],
    'Shares': [800, 1200, 400, 1500, 600],
    'Female_Percentage': [60, 55, 45, 65, 40],
    'Male_Percentage': [40, 45, 55, 35, 60]
})
```

```python
# Comparison Plots

# 1. Line Chart - Comparing Likes and Comments
plt.figure(figsize=(10, 6))
plt.plot(data['Platform'], data['Likes'], marker='o', label='Likes', color='blue')
plt.plot(data['Platform'], data['Comments'], marker='o', label='Comments', color='green')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.title('Likes vs Comments by Platform')
plt.legend()
plt.show()

# 2. Bar Chart - Engagement Metrics by Platform
plt.figure(figsize=(10, 6))
bar_width = 0.25
indices = np.arange(len(data))
plt.bar(indices, data['Likes'], width=bar_width, label='Likes')
plt.bar(indices + bar_width, data['Comments'], width=bar_width, label='Comments')
plt.bar(indices + 2 * bar_width, data['Shares'], width=bar_width, label='Shares')
plt.xticks(indices + bar_width, data['Platform'])
plt.xlabel('Platform')
plt.ylabel('Engagement Count')
plt.title('Engagement Metrics by Platform')
plt.legend()
plt.show()

# 3. Radar Chart - Engagement Metrics by Platform
categories = ['Likes', 'Comments', 'Shares']
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for i, row in data.iterrows():
    values = row[categories].tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=row['Platform'])
    ax.fill(angles, values, alpha=0.1)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title("Engagement Metrics by Platform")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
# Composition Plots

# 1. Pie Chart - Gender Composition on TikTok
plt.figure(figsize=(6, 6))
tiktok_gender = [data.loc[3, 'Female_Percentage'], data.loc[3, 'Male_Percentage']]
plt.pie(tiktok_gender, labels=['Female', 'Male'], autopct='%1.1f%%', colors=['pink', 'lightblue'])
plt.title('TikTok Gender Composition')
plt.show()

# 2. Donut Chart - Gender Composition on Instagram
plt.figure(figsize=(6, 6))
instagram_gender = [data.loc[0, 'Female_Percentage'], data.loc[0, 'Male_Percentage']]
plt.pie(instagram_gender, labels=['Female', 'Male'], autopct='%1.1f%%', colors=['pink', 'lightblue'], wedgeprops={'width': 0.3})
plt.title('Instagram Gender Composition')
plt.show()

# 3. Stacked Bar Chart - Engagement by Platform (Likes, Comments, Shares)
plt.figure(figsize=(10, 6))
plt.bar(data['Platform'], data['Likes'], label='Likes', color='skyblue')
plt.bar(data['Platform'], data['Comments'], bottom=data['Likes'], label='Comments', color='lightgreen')
plt.bar(data['Platform'], data['Shares'], bottom=data['Likes'] + data['Comments'], label='Shares', color='salmon')
plt.xlabel('Platform')
plt.ylabel('Engagement Count')
plt.title('Engagement by Platform')
plt.legend()
plt.show()

# 4. Stacked Area Chart - Cumulative Engagement by Platform
# Calculate cumulative data
cumulative_data = data[['Likes', 'Comments', 'Shares']].cumsum()
plt.figure(figsize=(10, 6))
plt.stackplot(data['Platform'], cumulative_data['Likes'], cumulative_data['Comments'], cumulative_data['Shares'],
              labels=['Likes', 'Comments', 'Shares'], colors=['skyblue', 'lightgreen', 'salmon'])
plt.title('Cumulative Engagement by Platform')
plt.xlabel('Platform')
plt.ylabel('Cumulative Engagement Count')
plt.legend(loc='upper left')
plt.show()
```
