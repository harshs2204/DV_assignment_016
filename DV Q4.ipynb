```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from CSV file
data = pd.read_csv('crop_yield.csv')

# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Ensure that we are working with a reasonable dataset size
if data.shape[0] > 10:
    data = data.sample(10)  # Sample 10 rows if too many for demonstration

# 1. Bar Plot for Crop Yield
plt.figure(figsize=(10, 6))
plt.bar(data['Crop'], data['Yield'], color='skyblue', label='Yield (tons/ha)')
plt.title('Crop Yield Comparison by Crop', fontsize=16)
plt.xlabel('Crops', fontsize=14)
plt.ylabel('Yield (tons/ha)', fontsize=14)

# Adding text on bars
for index, value in enumerate(data['Yield']):
    plt.text(index, value + 0.1, f'{value:.2f}', ha='center', color='blue')

plt.legend()
plt.grid(axis='y')  # Add grid for better readability
plt.tight_layout()
plt.show()
```

```python
# 2. Line Plot for Annual Rainfall
plt.figure(figsize=(10, 6))
plt.plot(data['Crop'], data['Annual_Rainfall'], marker='o', color='orange', label='Annual Rainfall (mm)', linewidth=2)

# Adding text on line plot
for index, value in enumerate(data['Annual_Rainfall']):
    plt.text(index, value + 20, f'{value}', ha='center', color='orange')

plt.title('Annual Rainfall Comparison by Crop', fontsize=16)
plt.xlabel('Crops', fontsize=14)
plt.ylabel('Annual Rainfall (mm)', fontsize=14)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Summarizing data by crop
production_summary = data.groupby('Crop')['Production'].sum().reset_index()

# Stacked Bar Chart for Production
plt.figure(figsize=(10, 6))
plt.bar(production_summary['Crop'], production_summary['Production'], color='lightgreen', label='Production (metric tons)')
plt.title('Production Comparison by Crop', fontsize=16)
plt.xlabel('Crops', fontsize=14)
plt.ylabel('Production (metric tons)', fontsize=14)

# Adding text on bars
for index, value in enumerate(production_summary['Production']):
    plt.text(index, value + 50, f'{value}', ha='center', color='darkgreen')

plt.legend()
plt.grid(axis='y')  # Add grid for better readability
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for Crop Yield Composition
plt.figure(figsize=(8, 8))
plt.pie(data['Yield'], labels=data['Crop'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Crop Yield Composition', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()

```

```python
# Bar Chart for Area by Crop
plt.figure(figsize=(10, 6))
plt.bar(data['Crop'], data['Area'], color='lightblue')
plt.title('Area Comparison by Crop', fontsize=16)
plt.xlabel('Crops', fontsize=14)
plt.ylabel('Area (hectares)', fontsize=14)

# Adding text on bars
for index, value in enumerate(data['Area']):
    plt.text(index, value + 50, f'{value}', ha='center', color='blue')

plt.grid(axis='y')  # Add grid for better readability
plt.tight_layout()
plt.show()

```
